### Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión lineal entre cada columna numérica con la columna 'price'. 

Cuando existe una columna(variable) que es la que se quiere comparar, se le llama 'target'. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python

df = pd.read_csv("../data/USA_Housing.csv")
df
```

```python
def funcion(columna):
    print("---------------------------------------")
    print("MOSTRANDO RELACION DE COLUMNA:=",columna)
    print("---------------------------------------")
    #Convirtiendo en vector (array)
    col_cambio_array = np.array(df[columna])
    #print(col_cambio_array)
    #print(type(col_cambio_array))
    #print(col_cambio_array.shape) #ahora mismo es un vector

    #Transformando a matriz
    col_cambio_array = col_cambio_array.reshape(-1, 1)
    #print(col_cambio_array)
    #print(type(col_cambio_array))
    #print(col_cambio_array.shape) #Ahora es una matriz

    X = col_cambio_array
    y = df["Price"]
    modelo = LinearRegression()
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20)
    modelo.fit(X_train, y_train)
    y_predecido = modelo.predict(X_test)


    print("el score de test =",modelo.score(X_test, y_test) * 100) 

    #Mostramos la correlacion que tiene. Es decir, demostración si mi modelo ha aprendido o no
    plt.scatter(X,y,edgecolors="black") 
    plt.plot()
    plt.show()
    print("MAE=", mean_absolute_error(y_test, y_predecido)) # para las casas esta fallando 216.537 dolares
    print("MSE=",mean_squared_error(y_test,y_predecido))
    print("RMSE=",np.sqrt(mean_squared_error(y_test,y_predecido)))
```

```python
funcion(columna="Avg. Area Income")
```

```python
funcion(columna="Avg. Area House Age")
```
