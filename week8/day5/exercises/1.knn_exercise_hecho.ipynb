```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/iris.csv")
df
```

- Mediante knn, predice qué planta se corresponde con los ejemplos: [15.7,2.8,9.5,0.1] y [1.7,3.8,4.5,11.3] para k=3 y k=7
- ¿Qué acierto tiene tu modelo para k=3, k=7? (score)

```python
X = df.drop(columns=["variety"])
y = df["variety"]
X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=3)
model3.fit(X_train,y_train)
print(model3.score(X_test,y_test))
#[15.7,2.8,9.5,0.1] y [1.7,3.8,4.5,11.3]
print(model3.predict([[15.7,2.8,9.5,0.1]]))
print(model3.predict([[1.7,3.8,4.5,11.3]]))
```

```python
model7 = KNeighborsClassifier(n_neighbors=7)
model7.fit(X_train,y_train)
print(model7.score(X_test,y_test))
#[15.7,2.8,9.5,0.1] y [1.7,3.8,4.5,11.3]
print(model7.predict([[15.7,2.8,9.5,0.1]]))
print(model7.predict([[1.7,3.8,4.5,11.3]]))
```

Entre 1 y 20, ¿qué k tiene un mejor acierto en tu modelo? visualiza el score con el valor de k en una gráfica:

```python
lista_de_scores = []
for i in range(1,21):
    model = KNeighborsClassifier(n_neighbors=i)
    entrenamiento = model.fit(X_train,y_train)
    scores = model.score(X_test, y_test)
    print("K = ",i,"Score=",scores)
    
    lista_de_scores.append(scores)

plt.scatter(range(1,21), lista_de_scores)
plt.xlabel('K')
plt.ylabel('Score')
```
